```python
from ultralytics import YOLO
import os

# COCO 원본 클래스 ID 유지한 경우에만 사용
KEEP_CLASSES = [25, 26, 39, 41, 63, 64, 66, 67, 73, 76, 78, 79]

if __name__ == "__main__":  # Windows 환경 안정성
    # 라벨 캐시 삭제
    for p in ["./dataset/labels/train.cache", "./dataset/labels/val.cache"]:
        if os.path.exists(p):
            os.remove(p)
            print("삭제:", p)

    model = YOLO("yolo26x-seg.pt")

    results = model.train(
        data="./dataset/data.yaml",
        epochs=30,                 # 15보다 조금 늘리고
        patience=8,                # 개선 없으면 조기 종료
        imgsz=640,
        batch=1,
        device="cpu",
        workers=0,

        optimizer="AdamW",
        lr0=0.0003,                # 기존보다 조금 더 보수적으로
        lrf=0.01,
        weight_decay=0.001,        # 과적합 방지
        warmup_epochs=3.0,
        cos_lr=True,               # 후반부 부드럽게 감소

        freeze=15,                 # 기존 클래스 유지 쪽으로 조금 더 보수적
        classes=KEEP_CLASSES,

        # augmentation: 너무 세지 않게
        hsv_h=0.015,
        hsv_s=0.5,
        hsv_v=0.3,
        translate=0.05,
        scale=0.2,
        fliplr=0.5,
        mosaic=0.5,
        close_mosaic=5,
        mixup=0.0,
        cutmix=0.0,
        copy_paste=0.15,           # segment 전용, 약하게만
        overlap_mask=False,        # 이전 mask 에러가 있었으면 유지

        seed=42,
        deterministic=True,
        val=True,
        plots=True,
        save=True,
        save_period=5,

        project="runs",
        name="seg_yolo26x_fusion_main_final",
        exist_ok=True
    )
```
